```python
url = "https://raw.githubusercontent.com/Apress/mastering-ml-w-python-in-six-steps/master/Chapter_3_Code/Code/Data/Grade_Set_1.csv"
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(url)
```

```python
df
```

```python
x=df['Hours_Studied'].values
x = x.reshape(-1,1)
x.shape
x
```

```python
y=df['Test_Grade'].values
y=y.reshape(-1,1)
y.shape
```

```python
model = LinearRegression().fit(x,y)
```

```python
r_sq=model.score(x,y)
print('coefficients of determination: ', r_sq)
```

```python
print("intercept", model.intercept_)
```

```python
print("slope", model.coef_)
```

```python
plt.scatter(x,y, color='blue')
plt.plot(x, model.predict(x), color='red')
plt.title('korrelation studietimmar och poäng')
plt.xlabel('Studietimmar')
plt.ylabel('Poäng')
plt.show()
```
